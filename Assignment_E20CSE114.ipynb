```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
excel_file = r'C:\Users\Nec\Documents\Assignment_E20CSE114.xlsx'
data = pd.read_excel(excel_file)
print(data.head())
```

```python
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(df['Date'], df['Price'], marker='o', linestyle='-')

# Formatting the date axis
plt.gcf().autofmt_xdate()  # Auto-formatting date display

# Labeling axes and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Date vs. Price')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['10_day_MA'] = df['Price'].rolling(window=10).mean() #10-day moving average
df['50_day_MA'] = df['Price'].rolling(window=50).mean() #50-day moving average
plt.plot(df['Date'], df['Price'], label='Original Data', marker='o', linestyle='-')

# Plot 10-day moving average
plt.plot(df['Date'], df['10_day_MA'], label='10-day Moving Average', linestyle='--')

# Plot 50-day moving average
plt.plot(df['Date'], df['50_day_MA'], label='50-day Moving Average', linestyle='--')

# Formatting the date axis
plt.gcf().autofmt_xdate()  # Auto-formatting date display

# Labeling axes and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Original Data with Moving Averages')
plt.legend()  # Show legend

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
